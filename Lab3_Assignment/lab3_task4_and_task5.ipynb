```python
#task 4 simple example
#code used from UCS example
dd = {'s':[('a',3),('b',2),('c',4)],
    'a':[('d',1)],
    'b':[('e',8),('f',7),('d',1)],
    'c':[('e',1),('j',3)],
    'd':[('f',1),('h',1)],
    'e':[('g',1)],
    'f':[('g',5)],
    'h':[('g',1),('f',3)],
    'j':[('e',2)],
    'g':[]}

def build_dcit(lis):
    di  = {}

    for i in lis:
        name = i[0]
        cost = i[1]
        di[name] = cost
    return di

def my(x):
    return x[1]
```

```python
def UCS(d,goal):
    
    q = []
    q.append(('s',0))
    explored = []
    
    while len(q)>0:

        print('Queue :  ',q)
        node = q.pop(0)
        print('child pop: ',node) # node[0] --> node name AND node[1] --> cost
        
        if node[0] not in explored:
            explored.append(node[0])
        
        if node[0] == goal:
            return explored 
        
        print('Explored: ',explored)

        child = d.get(node[0])
        print('current child: ',child)
        
        for i in child:
            print(i)
            
            cost =  i[1] + node[1]
            name = i[0]
            new = (name,cost)

            
            print('{} , parent cost ({}) + child cost ({}) = {}'.format(i,node[1],i[1],cost))
            
            
            if new[0] not in explored and new[0] not in build_dcit(q):
                
                q.append(new) 
                
            
                
            elif new[0] in build_dcit(q):


                old_data = build_dcit(q)
                old_cost = old_data[new[0]]
                old_name = new[0]
                old = (old_name,old_cost)


                if new[1] < old[1]:
                    inde = q.index(old)
                    q[inde] = new
                else:
                    pass

        
        q = sorted(q,key= my)
        #print(q)
        print(50*'*')
        
    
result=UCS(dd,'g')#UCS is like BFS

print("total nodes explored :",result)
```
