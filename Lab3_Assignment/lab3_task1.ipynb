```python
#graph
from collections import defaultdict
class Graph:
    def __init__(self):
        self.graph=defaultdict(list)
        
    def addEdge(self,v,u):
        self.graph[u].append(v)
        
    def BFS(self,s):
        visited=[False]*(max(self.graph)+1) # makes all of the vertices as not visited
        
        q=[]
        q.append(s)# insert the first element
        visited[s]=True
        
        while q:
            s=q.pop(0)
            print(s,end=" ")
            
            for i in self.graph[s]:
                if visited[i]==False:
                    q.append(i)
                    visited[i]=True
    
    def DFSUtil(self, v, visited):# recursive function
 
        # Mark the current node as visited
        # and print it
        visited.add(v)
        print(v, end=' ')
 
        # Recur for all the vertices
        # adjacent to this vertex
        for neighbour in self.graph[v]:
            if neighbour not in visited:
                self.DFSUtil(neighbour, visited)
 
    def DFS(self, v):
 
        # Create a set to store visited vertices
        visited = set()
 
        # Call the recursive helper function
        # to print DFS traversal
        self.DFSUtil(v, visited)#calling since we have to use it recrusivly
 
    
                    
g=Graph()
g.addEdge(0,1)
g.addEdge(1, 2)
g.addEdge(2, 0)
g.addEdge(2, 3)
g.addEdge(3,2)
g.addEdge(3,4)
g.addEdge(4,1)


g.BFS(2)
print("")

g.DFS(2)
```
